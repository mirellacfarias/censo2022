```python
# Função para desfazer células mescladas
def desmesclar_celulas(arquivo, novo_arquivo):
    workbook = openpyxl.load_workbook(arquivo)
    sheet = workbook.active
    
    # Loop para desfazer células mescladas
    merged_cells = list(sheet.merged_cells)
    for merged_cell in merged_cells:
        min_col, min_row, max_col, max_row = merged_cell.min_col, merged_cell.min_row, merged_cell.max_col, merged_cell.max_row
        top_left_cell_value = sheet.cell(row=min_row, column=min_col).value
        
        sheet.unmerge_cells(start_row=min_row, start_column=min_col, end_row=max_row, end_column=max_col)
        
        for row in range(min_row, max_row + 1):
            for col in range(min_col, max_col + 1):
                sheet.cell(row=row, column=col, value=top_left_cell_value)
    
    workbook.save(novo_arquivo)
    print(f"Arquivo salvo como: {novo_arquivo}")
```

```python
# Processar cada arquivo
for nome, caminho in arquivos.items():
    novo_caminho = rf"{pasta_desmesclados}\{nome}_dm.xlsx"
    desmesclar_celulas(caminho, novo_caminho)

    # Carregar o arquivo modificado com pandas
    df = pd.read_excel(novo_caminho)
    print(f"Dados de {nome}:")
    print(df.head())  # Exibe as primeiras linhas do dataframe para verificação
    print("\n")
```

```python
# Caminho dos arquivos desmesclados
arquivos_dm = {
    "Amarelos": r"C:\Users\mirel\Documents\Projetos_Mirella\Mirella-Farias\Censo2022\Arquivos_Desmesclados\Amarelos_dm.xlsx",
    "Brancos": r"C:\Users\mirel\Documents\Projetos_Mirella\Mirella-Farias\Censo2022\Arquivos_Desmesclados\Brancos_dm.xlsx",
    "Indigenas": r"C:\Users\mirel\Documents\Projetos_Mirella\Mirella-Farias\Censo2022\Arquivos_Desmesclados\Indigenas_dm.xlsx",
    "Pardos": r"C:\Users\mirel\Documents\Projetos_Mirella\Mirella-Farias\Censo2022\Arquivos_Desmesclados\Pardos_dm.xlsx",
    "Pretos": r"C:\Users\mirel\Documents\Projetos_Mirella\Mirella-Farias\Censo2022\Arquivos_Desmesclados\Pretos_dm.xlsx"
}
```

```python
# Função para carregar e transformar os dados
def carregar_transformar_dados(arquivo, raca):
    # Carregar dados a partir da linha 4429 ignorando a última linha
    df = pd.read_excel(arquivo, skiprows=4428, skipfooter=1)
    
    # Listar colunas para inspeção
    print(df.columns)
    
    # Assumindo que as colunas são as mesmas em todos os arquivos, ajustar conforme necessário
    col_nivel_dado = df.columns[0]
    col_municipio = df.columns[1]
    col_idade = df.columns[2]
    col_homens = df.columns[4]
    col_mulheres = df.columns[5]
        
    # Filtrar apenas os dados municipais
    df = df[df[col_nivel_dado] == 'MU']
    
    # Separar município e estado
    df[['Município', 'UF']] = df[col_municipio].str.extract(r'(.+?)\s+\((.+?)\)')
    
    # Reorganizar colunas e derreter os dados
    df = df[['Município', 'UF', col_idade, col_homens, col_mulheres]]
    df = df.melt(id_vars=['Município', 'UF', col_idade], value_vars=[col_homens, col_mulheres], 
                 var_name='Sexo', value_name='Pessoas')
    
    # Mapear nomes das colunas originais para novos valores de "Sexo"
    sexo_map = {col_homens: 'Homens', col_mulheres: 'Mulheres'}
    df['Sexo'] = df['Sexo'].map(sexo_map)
    
    # Adicionar coluna de raça
    raca_map = {
        "Amarelos": "Amarela",
        "Brancos": "Branca",
        "Indigenas": "Indígena",
        "Pardos": "Parda",
        "Pretos": "Preta"
    }
    df['Raça'] = raca_map[raca]
    
    # Reordenar colunas
    df = df[['Município', 'UF', col_idade, 'Raça', 'Sexo', 'Pessoas']]
    df.columns = ['Município', 'UF', 'Idade', 'Raça', 'Sexo', 'Pessoas']
    
    return df
```

```python
# Carregar e transformar todos os dados
dados_completos = []
for raca, caminho in arquivos_dm.items():
    df_transformado = carregar_transformar_dados(caminho, raca)
    dados_completos.append(df_transformado)

# Concatenar todos os dados em um único DataFrame
dados_completos_df = pd.concat(dados_completos, ignore_index=True)

# Função para salvar dados em múltiplas planilhas se necessário
def salvar_em_multiplas_planilhas(df, caminho_base, max_linhas=1048575):
    num_linhas = len(df)
    num_planilhas = (num_linhas // max_linhas) + 1
    for i in range(num_planilhas):
        inicio = i * max_linhas
        fim = (i + 1) * max_linhas
        df_subset = df.iloc[inicio:fim]
        caminho = f"{caminho_base}_parte{i+1}.xlsx"
        df_subset.to_excel(caminho, index=False)

# Salvar dados completos em múltiplas planilhas
salvar_em_multiplas_planilhas(dados_completos_df, r"C:\Users\mirel\Documents\Projetos_Mirella\Mirella-Farias\Censo2022\Dados_Completos")

# Exibir uma amostra dos dados transformados
print(dados_completos_df.head())
```
